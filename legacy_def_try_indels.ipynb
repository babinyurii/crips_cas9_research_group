```python
from Bio import SeqIO
import copy
import pandas as pd
```

```python
def count_indels(input_file):
    
    
    for record in SeqIO.parse(input_file, "fasta"):
        ref_seq = record
        break
        
    total_insertions = [[] for x in range(len(ref_seq))]
    total_deletions = [[] for x in range(len(ref_seq))]
    
    reads = SeqIO.parse(input_file, "fasta")

    
    for read in reads:
        
        if read.id == ref_seq.id:
            continue
        
        start_nuc = 0
        for nuc in range(len(read)):
            if nuc < start_nuc:
                continue
            
            if ref_seq[nuc] != read[nuc]:
                
                if read[nuc] == "-":
                    
                    deletions_counter = 0
                    start_nuc = 0
                    
                    for nuc_del in range(nuc + 1, len(read)):
                        if read[nuc_del] == "-" and ref_seq[nuc_del] != read[nuc_del]:
                            deletions_counter += 1
                        if read[nuc_del] != "-":
                            start_nuc = nuc_del
                            break
                    total_deletions[nuc_del].append(deletions_counter + 1)
                
                
                if ref_seq[nuc] == "-" and read[nuc] != "N":
                    insertion_counter = 0
                    start_nuc = 0
                    
                    for nuc_ins in range(nuc + 1 , len(read)):
                        if ref_seq[nuc_ins] == "-" and read[nuc_ins] != ref_seq[nuc_ins]:
                            insertion_counter += 1
                        if ref_seq[nuc_ins] != "-":
                            start_nuc = nuc_ins
                            break
                    total_insertions[nuc_ins].append(insertion_counter + 1)


        
    print(total_deletions)
    print(total_insertions)
    
```
